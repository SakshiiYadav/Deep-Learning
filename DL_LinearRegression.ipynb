Regression using Feed Forward Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense #Feed forward NN
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)

X = np.random.rand(1000,3)  #Vectors will be generated
X[0:5]
```

```python
def true_func(X):
  return np.sin(1.5*np.pi*X[:,0]) + np.cos(1.5*np.pi*X[:,1]) + 2*X[:,2]
```

```python
y = true_func(X)
```

```python
y[0:5]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
model = Sequential()
```

```python
model.add(Dense(15,input_dim=3,activation ='relu'))
model.add(Dense(1,activation='linear'))             #linear is same as identity function
```

```python
model.compile(loss='mean_squared_error',optimizer='adam',metrics='mean_absolute_error')
```

```python
model.fit(xtrain,ytrain,epochs=150,batch_size=30)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
predictions =model.predict(xtest)
```

```python
mean_absolute_error(ytest,predictions)
```

## Make a dataset with 3 columns 'mass', 'accelerate' and 'force' where mass and acceleration will be input features and force will be the target to be predicted.
## Design a nn which takes mass and acceleration as input and predicts corresponding force (Force = mass*acceleration)

```python
import pandas as pd
x =pd.DataFrame(np.random.randint(0,10,size=(20,2)), columns=['mass','acceleration'])
x
```

```python
x['Force'] = x.mass*x.acceleration
x
```

```python
x = x.iloc[:20,:2]
y = x.iloc[:20, -1]
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
model = Sequential()
```

```python
model.add(Dense(13,input_dim=2,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.compile(loss='mean_squared_error',optimizer='adam',metrics='mean_absolute_error')
```

```python
model.fit(xtrain,ytrain,epochs=250,batch_size=32)
```

```python
predictions = model.predict(xtest)
```

```python
mean_absolute_error(ytest,predictions)
```
