# Multiclass classification on Wine Dataset

```python
import pandas as pd
from sklearn.datasets import load_wine
from keras.models import Sequential
from keras.layers import Dense
```

```python
data = load_wine()
```

```python
df = pd.DataFrame(data.data)
```

```python
df
```

```python
df.shape
```

```python
X = pd.DataFrame(data.data)
y = pd.DataFrame(data.target)
```

```python
X
```

```python
y.head()
```

```python
y.tail()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from keras.utils import np_utils
encoded_Y = np_utils.to_categorical(y_train)
encoded_Y
```

```python
model=Sequential()
model.add(Dense(13,input_dim=13,activation='relu'))
model.add(Dense(14,activation='relu'))
model.add(Dense(3,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics='accuracy')
```

```python
model.fit(X_train,encoded_Y,epochs=100,batch_size=10)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
def vec2scalar(list_):
    pred = []
    for i in list_:
        val = [round(j,0) for j in i]
        if val[0]==1.0:
            pred.append(0)
        elif val[1]==1.0:
            pred.append(1)
        else:
            pred.append(2)
    return pred
```

```python
print(vec2scalar(y_pred))
```
